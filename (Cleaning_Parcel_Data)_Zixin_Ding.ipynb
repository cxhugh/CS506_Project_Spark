```python
# import cell
import pandas as pd
import numpy as np

data14 = pd.read_csv("property-assessment-fy2014.csv")

data15 = pd.read_csv("property-assessment-fy2015.csv")

data16 = pd.read_csv("property-assessment-fy2016.csv")
#data16 = data16[['PID', 'R_HEAT_TYP']]


data17 = pd.read_csv("property-assessment-fy2017.csv")
#data17 = data17[['PID', 'R_HEAT_TYP']]


data18 = pd.read_csv("ast2018full.csv")
#data18 = data18[['PID', 'R_HEAT_TYP']]
```

```python
#selecting only interested column
train14 = data14[['Parcel_ID', 'ZIPCODE', 'R_HEAT_TYP', 'LU', 'AV_TOTAL',
                 "YR_BUILT", "STRUCTURE_CLASS","R_AC",
                 "U_AC", "U_HEAT_TYP", "OWN_OCC", "LIVING_AREA"]]
train15 = data15[['PID', 'ZIPCODE', 'R_HEAT_TYP', 'LU', 'AV_TOTAL',
                 "YR_BUILT", "STRUCTURE_CLASS","R_AC",
                 "U_AC", "U_HEAT_TYP", "OWN_OCC", "LIVING_AREA"]]
train16 = data16[['PID', 'ZIPCODE', 'R_HEAT_TYP', 'LU', 'AV_TOTAL',
                 "YR_BUILT", "STRUCTURE_CLASS","R_AC",
                 "U_AC", "U_HEAT_TYP", "OWN_OCC", "LIVING_AREA"]]
train17 = data17[['PID', 'ZIPCODE', 'R_HEAT_TYP', 'LU', 'AV_TOTAL',
                 "YR_BUILT", "STRUCTURE_CLASS","R_AC",
                 "U_AC", "U_HEAT_TYP", "OWN_OCC", "LIVING_AREA"]]
train18 = data18[['PID', 'ZIPCODE', 'R_HEAT_TYP', 'LU', 'AV_TOTAL',
                 "YR_BUILT", "STRUCTURE_CLASS","R_AC",
                 "U_AC", "U_HEAT_TYP", "OWN_OCC", "LIVING_AREA"]]
```

```python
train14.head(10)
# First, by looking at data, I found R_HEAT_TYP and U_HEAT_TYP don't coexist.
# the dependent variable of our final model is Heat type

#drop rows when R_HEAT_TYP and U_HEAT_TYP are both NAN
train14 = train14.dropna(subset=["R_HEAT_TYP", "U_HEAT_TYP"], how='all')
train15 = train15.dropna(subset=["R_HEAT_TYP", "U_HEAT_TYP"], how='all')
train16 = train16.dropna(subset=["R_HEAT_TYP", "U_HEAT_TYP"], how='all')
train17 = train17.dropna(subset=["R_HEAT_TYP", "U_HEAT_TYP"], how='all')
train18 = train18.dropna(subset=["R_HEAT_TYP", "U_HEAT_TYP"], how='all')
```

```python
train14.head(15)
# since we want to do a regression model, we have to delete rows where AV_TOTAL is zero
# AV_TOTAL: Total assessed value for property
train14 = train14[(train14.AV_TOTAL != 0)]
train15 = train15[(train15.AV_TOTAL != 0)]
train16 = train16[(train16.AV_TOTAL != 0)]
train17 = train17[(train17.AV_TOTAL != 0)]
train18 = train18[(train18.AV_TOTAL != 0)]
```

```python
train14.ZIPCODE = train14.ZIPCODE.str.rstrip('_')
train15.ZIPCODE = train15.ZIPCODE.str.rstrip('_')
train16.ZIPCODE = train16.ZIPCODE.str.rstrip('_')
train17.ZIPCODE = train17.ZIPCODE.str.rstrip('_')
train18.ZIPCODE = train18.ZIPCODE.astype(int).astype(str).str.zfill(5)

#deal with zipcode
```

```python
train18.head(25)
```

```python
pd.isnull(train14.loc[2]["U_AC"])
```

```python
listtrain = [train14, train15, train16, train17, train18]
dflist = []
for i in range(len(listtrain)):
    a = listtrain[i]
    #a = a.fillna("N/A")
    a.RorU = "R"
    a.AC = "C"
    a.HEAT_TYP = ""
    for i, row in a.iterrows():  
        RorU_val = "U"
        AC_val = "N"
        HEAT_TYP_val = "N"
    
        if pd.isnull(a.loc[i]['R_AC']) == False:
            RorU_val = "R"
            AC_val = a.loc[i]["R_AC"]
            HEAT_TYP_val = a.loc[i]["R_HEAT_TYP"]
        else:
            RorU_val = "U"
            AC_val = a.loc[i]["U_AC"]
            HEAT_TYP_val = a.loc[i]["U_HEAT_TYP"]
        
        a.set_value(i, "RorU", RorU_val)
        a.set_value(i, "AC", AC_val)
        a.set_value(i, "HEAT_TYP", HEAT_TYP_val)

    a = a.drop(['R_AC', 'U_AC', 'R_HEAT_TYP', 'U_HEAT_TYP'], axis=1)
    dflist.append(a) 


```

```python
# traindata together in a list
dflist[0]
dflist[0].columns = ['PID' if x=='Parcel_ID' else x for x in dflist[0].columns]
dflist[0].head(20)
```

```python
for i in range(len(dflist)):
    print(dflist[i].LU.unique())
# see how many levels of the Type of property column in each data set
    dflist[i]["Year"] = str(i+2014)
```

```python
import copy
finaldata = copy.deepcopy(dflist[0])
finaldata = finaldata.append([dflist[1], dflist[2], dflist[3], dflist[4]])
finaldata.shape
```

```python
finaldata.loc[finaldata["HEAT_TYP"] == 'w', "HEAT_TYP"] = "W"
finaldata.loc[finaldata["HEAT_TYP"] == ' F', "HEAT_TYP"] = "F"
finaldata.loc[finaldata["HEAT_TYP"] == 'f', "HEAT_TYP"] = "F"
finaldata.HEAT_TYP.unique()
# misrecoding adjustification
```

```python
# delete rows where HEAT_TYP equal 'D', 'Y', 'C', 'N/A', ' ', 'I','G'
finaldata = finaldata.drop(finaldata[finaldata["HEAT_TYP"].isin(['D', 'Y', 'C', ' ', 'I','G'])].index)
finaldata = finaldata.dropna(subset=["HEAT_TYP"]) 
print(finaldata.shape)
finaldata.head(20)
```

```python
print(finaldata.HEAT_TYP.unique())

# delete 'XX' 'CC' 'C' in LU column CC and C commercial XX, and 102 don't exist in key misrecording
finaldata = finaldata.drop(finaldata[finaldata["LU"].isin(['XX', 'CC', 'C', '102'])].index)
print(finaldata.shape)
print(finaldata.LU.unique())
```

```python
finaldata.head(20)

finaldata.YR_BUILT = pd.to_numeric(finaldata.YR_BUILT, errors = 'coerce')
finaldata.Year = pd.to_numeric(finaldata.Year, errors = 'coerce')
finaldata.LIVING_AREA = pd.to_numeric(finaldata.LIVING_AREA, errors = 'coerce')
print(finaldata.dtypes)
```

```python
#drop all na in every column
finaldata = finaldata.dropna()
finaldata.shape
```
